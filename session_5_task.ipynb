###`Q1:` Join Tuples if similar initial element
While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the similarity of initial element. This problem can have applications in data domains such as Data Science.

For eg.
```
Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
Output : [(5, 6, 7, 8), (6, 10), (7, 13)]
```

````python
# write your code here
test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
d={}

for i in range(len(test_list)):
 if test_list[i][0] in d.keys():
  d[test_list[i][0]] = d[test_list[i][0]] + " "+str(test_list[i][1])

 else:
   d[test_list[i][0]] =str(test_list[i][1])

result=[]
for key,value in d.items():
  values=value.split(" ")
  # print(type(values))
  values=list(map(int,values))
  result.append((key,values))

print(result)
````

###`Q2:` Multiply Adjacent elements (both side) and take sum of right and lest side multiplication result.


For eg.
```
The original tuple : (1, 5, 7, 8, 10)
Resultant tuple after multiplication :

(1*5, 1*5+5*7, 7*5 + 7*8, 8*7 + 8*10, 10*8) -> (5, 40, 91, 136, 80)

output-(5, 40, 91, 136, 80)
```

````python
# write your code here
test_list = (1, 5, 7, 8, 10)
result=[]
for i in range(len(test_list)):
  if i==0:
    result.append(test_list[i])
  elif i==len(test_list)-1:
    result.append(test_list[i]*test_list[i-1])
  else:
      result.append(test_list[i]*test_list[i-1]+test_list[i]*test_list[i+1])

print(result)
````

###`Q3`: Check is tuples are same or not?
Two tuples would be same if both tuples have same element at same index
```
t1 = (1,2,3,0)
t2 = (0,1,2,3)

t1 and t2 are not same
```

````python
# write your code here
t1 = (1,2,3,0)
t2 = (1,2,3,0)

for i in range(len(t1)):
  if t1[i]!=t2[i]:
    print("t1 and t2 are not same")
    break
  elif len(t1)!=len(t2):
    print("t1 and t2 are not same")
else:
  print("t1 and t2 are same")
````

###`Q4`: Count no of tuples, list and set from a list
```
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

```
`Output:`

```
List-2
Set-2
Tuples-1
```

````python
# write your code here
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]
d={'list':0,'tuple':0,'set':0}
for i in list1:
  if type(i)==list:
    if d['list']==0:
      d['list']=1
    else:
      d['list']=d['list']+1
  elif type(i)==set:
     if d['set']==0:
      d['set']=1
     else:
      d['set']=d['set']+1
  elif type(i)==tuple:
     if d['tuple']==0:
      d['tuple']=1
     else:
      d['tuple']=d['tuple']+1

for k,v in d.items():
  print(k,"->",v)
````

###`Q1:` Write a program to find set of common elements in three lists using sets.
```
Input : ar1 = [1, 5, 10, 20, 40, 80]
        ar2 = [6, 7, 20, 80, 100]
        ar3 = [3, 4, 15, 20, 30, 70, 80, 120]

Output : [80, 20]
```

````python
# write your code here
ar1 = [1, 5, 10, 20, 40, 80]
ar2 = [6, 7, 20, 80, 100]
ar3 = [3, 4, 15, 20, 30, 70, 80, 120]
result=[]
for i in ar1:
  if i in ar2 and i in ar3:
    result.append(i)
print(result)
#set way
ar1=set(ar1)
ar2=set(ar2)
ar3=set(ar3)
print(list(ar1.intersection(ar2,ar3)))
````
